```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Type of plots 

#### Line plots

```python
x = [1,2,3,4,5]
y=[10,20,25,30,40]
```

## Bar Chart

```python
months= ['Jan', 'Feb', 'Mar', 'Apr']
revenue=[100,200,300,500]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months, revenue)
plt.title('simple line plot');
plt.xlabel('X-axis');
plt.ylabel('Y-axis');
```

## Subplots and Legends

```python
days=[1,2,3,4,5]
original_price=[10,20,25,30,40]
predicated_price=[8,21,25,28,38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicated_price,'r-o', label='Predicted Price') 
plt.legend()
plt.title('plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

## Subplots

```python
# First subplot- Original price
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.legend()
plt.title('original price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');

# second subplot- Predicated price
plt.subplot(3,1,2)
plt.plot(days, predicated_price,'r-o', label='Predicted Price')
plt.legend()
plt.title('Predicated price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');

# Third subplot- difference b/w Original & Predicated price 
price_diff=[o-p for o, p in zip(original_price, predicated_price)]
plt.subplot(3,1,3)
plt.bar(days, price_diff,color='purple', label='Price differnce')
plt.legend()
plt.title('Difference between original & Predicated Prices');
plt.xlabel('days');
plt.ylabel('Price difference');

plt.tight_layout()
```

# saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicated_price,'r-o', label='Predicted Price') 
plt.legend()
plt.title('plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.savefig('example-prediction.png',dpi=300) #saving
```
